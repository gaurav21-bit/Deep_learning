```python
tokens = simple_preprocess(text)
```

```python
sentences = [tokens[i:i+20] for i in range(0, len(tokens), 20)]
```

```python
# Train CBOW model (sg=0)
cbow_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    sg=0  # CBOW
)
```

```python
# Train Skip-Gram model (sg=1)
skipgram_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    sg=1  # Skip-Gram
)
```

```python
#  Vocabulary size
print("CBOW Vocabulary Size:", len(cbow_model.wv))
print("Skip-Gram Vocabulary Size:", len(skipgram_model.wv))
```

```python
# Display word vectors (5 words)
words = ["artificial", "data", "learning", "technology", "future"]
```

```python
print("\n--- CBOW Word Vectors ---")
# Use words that are present in the model's vocabulary
words_in_vocab = ['constitution', 'government', 'rights', 'justice', 'equality']
for word in words_in_vocab:
    if word in cbow_model.wv:
        print(f"{word}:\n", cbow_model.wv[word][:10])  # first 10 values
    else:
        print(f"Warning: '{word}' not found in CBOW model vocabulary.")
```

```python
print("\n--- Skip-Gram Word Vectors ---")
# Use words that are present in the model's vocabulary
words_in_vocab = ['constitution', 'government', 'rights', 'justice', 'equality']
for word in words_in_vocab:
    if word in skipgram_model.wv:
        print(f"{word}:\n", skipgram_model.wv[word][:10])  # first 10 values
    else:
        print(f"Warning: '{word}' not found in Skip-Gram model vocabulary.")
```

```python
#Find similar words
target_word = "constitution"

print("\nTop 5 similar words (CBOW):")
print(cbow_model.wv.most_similar(target_word, topn=5))

print("\nTop 5 similar words (Skip-Gram):")
print(skipgram_model.wv.most_similar(target_word, topn=5))
```
